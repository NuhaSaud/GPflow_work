This notebook shows how to detect numerical errors such as failure of Cholesky decomposition or failure of optimization and how to test for bad solution. Bad solution is detected when the GP posterior predictive is higher or lower than observations. To fix numerical errors or bad solution we fit a GP with new random hyper-parameters.

```python
import os
import numpy as np
import tensorflow as tf
import gpflow
import pandas as pd
import warnings
from sklearn import metrics
from scipy.signal import savgol_filter
import scipy.stats as ss
from matplotlib import pyplot as plt
import statsmodels.api as sm
warnings.filterwarnings('ignore')

%pylab inline 
```

```python
filename = 'wt_normalized_counts_scaled.csv'
Y = pd.read_csv(filename,index_col=[0])
selected_genes = ['SPAC56F8.12','SPAC1F5.02','SPBC16A3.18']
Y = Y.loc[selected_genes]
genes_name = Y.index.values.tolist() # genes name
Y = Y.values.astype(int) # gene expression matrix
X = pd.read_csv('wt_col_data_scaled.csv',index_col=[0])
X = X[['minute']]
X = X.values.astype(float) # time points 
X = X.reshape([-1,1])
```

```python
def plot(likelihood,model,mean,var):
    fig = plt.figure()
    plt.tick_params(labelsize='large', width=2)     
    plt.ylabel('Gene Expression', fontsize=16)
    plt.xlabel('Times', fontsize=16)
    c = 'royalblue'
    xtest = np.linspace(np.min(X)-.1,np.max(X)+.1,100)[:,None]
    plt.plot(xtest, mean, lw=2) 
    plt.scatter(model.data[0],model.data[1],s=30,marker = 'o',color=c,alpha=1.)
    
    if likelihood == 'Gaussian':
        plt.fill_between(xtest[:,0],
                            mean[:,0] - 1*np.sqrt(var[:,0]),
                            mean[:,0] + 1*np.sqrt(var[:,0]),color=c,alpha=0.2) # one standard deviation
        plt.fill_between(xtest[:,0],
                            mean[:,0] - 2*np.sqrt(var[:,0]),
                            mean[:,0] + 2*np.sqrt(var[:,0]),color=c, alpha=0.1)# two standard deviation
    else:
       
        lowess = sm.nonparametric.lowess    
        # one standard deviation 68%
        percentile_16 = lowess(np.percentile(var, 16, axis=0),xtest[:,0],frac=1./5, return_sorted=False)
        percentile_16 = [(i > 0) * i for i in percentile_16]
        percentile_84 = lowess(np.percentile(var, 84, axis=0),xtest[:,0],frac=1./5, return_sorted=False)
        percentile_84 = [(i > 0) * i for i in percentile_84]
        plt.fill_between(xtest[:,0],percentile_16,percentile_84,color=c,alpha=0.2)
        
        # two standard deviation 95%
        percentile_5 = lowess(np.percentile(var, 5, axis=0),xtest[:,0],frac=1./5, return_sorted=False)
        percentile_5 = [(i > 0) * i for i in percentile_5]
        percentile_95 = lowess(np.percentile(var,95, axis=0),xtest[:,0],frac=1./5, return_sorted=False)
        percentile_95 = [(i > 0) * i for i in percentile_95]
        plt.fill_between(xtest[:,0],percentile_5,percentile_95,color=c,alpha=0.1)
       
    plt.show()
    
def samples_posterior_predictive_distribution(xtest,model): # Poisson likelihood
        
        var = []
        f_samples = []
        for i in range(20):
            f_samples.append(model.predict_f_samples(xtest, 10))
            f = np.vstack(f_samples)
            link_f = np.exp(f[:, :, 0])
            y = []
            link_f_mean = np.mean(link_f, 0)
            for i in range(link_f_mean.shape[0]):
                y.append(ss.poisson.rvs(link_f_mean[i], size = 500)) 
            y = np.vstack(y)
            var.append(y.T)

        var = np.vstack(var)
        mean = savgol_filter(np.mean(var,axis = 0), int(xtest.shape[0]/2)+1, 3) # mean smoothing
        mean = [(i > 0) * i for i in mean] # remove negative values
        return mean,var
    
# Hyper-parameters initialization
def init_hyper_parameters(seed_value):
    hyper_parameters = {}   
    hyper_parameters['ls'] = np.random.uniform(0. , 100.)    
    hyper_parameters['var'] = np.random.uniform(1., 20.)

    tf.compat.v1.get_default_graph()
    tf.compat.v1.set_random_seed(seed_value)
    tf.random.set_seed(seed_value)
    gpflow.config.set_default_float(np.float64)
    
    return hyper_parameters

# fit a GP and fix Cholesky decomposition failure,optimization failure and bad solution 
# by random initialization if detected  

def fit_GP(count_fix,lik_name):
    np.random.seed(count_fix)
    hyper_parameters = init_hyper_parameters(count_fix)
    fit_successed = True   
    try:
        kern = gpflow.kernels.RBF( variance= hyper_parameters['var'],lengthscales = hyper_parameters['ls'])     
        
        y = Y[i].astype(float)
        y = y.reshape([-1,1])
        
        if lik_name == 'Gaussian':
            y = np.log(y+1)
            model = gpflow.models.GPR((X,y), kern)
        else:
            model = gpflow.models.VGP((X,y), kernel= kern,likelihood= gpflow.likelihoods.Poisson())
        
        o = gpflow.optimizers.Scipy()
        res = o.minimize(model.training_loss, variables=model.trainable_variables,options=dict(maxiter=5000))
        
        if not(res.success):
            if count_fix < 10:
                print('Optimization fail, fix it by random initialization.')
                count_fix = count_fix +1 
                fit_successed,model = fit_GP(count_fix,lik_name)
            else:
                print('Can not fit a Gaussian process, Optimization fail.')
                fit_successed = False
    except tf.errors.InvalidArgumentError as e:
        if count_fix < 10:
            print('Cholesky decomposition was not successful, fix it by random initialization.')
            count_fix = count_fix +1 
            fit_successed,model = fit_GP(count_fix,lik_name)
        else:
            print('Can not fit a Gaussian process, Cholesky decomposition was not successful.')
            fit_successed = False
            
    # check bad solution when the  mean and variance of a new data points
    # are above or blew data points with certian percentage
    if fit_successed:
        xtest = np.linspace(np.min(X),np.max(X),100)[:,None]
        if lik_name == 'Gaussian':
            mean, var = model.predict_y(xtest)
            mean = mean.numpy()
            var = var.numpy()
        else:
            # mean and varicance of samples from posterior predictive distribution 
            mean,var = samples_posterior_predictive_distribution(xtest,model)         
        
        if count_fix < 5: # limit number of trial to fix bad solution 
            y_mean = np.mean(y)
            mean_mean = np.mean(mean)
            y_max = np.max(y)
            mean_max = np.max(mean)
            y_min = np.abs(np.min(y))
            mean_min = np.abs(np.min(mean))
            
            if y_mean > 0.0 and (mean_max > y_max or mean_min < y_min):
                if abs(round((mean_mean-y_mean)/y_mean)) > 0 or mean_mean == 0.0:
                    print('bad solution, fix it by random initialization.')
                    plot(lik_name,model,mean,var)
                    count_fix = count_fix + 1 
                    fit_successed,model = fit_GP(count_fix,lik_name)

    return fit_successed,model

def fit_model(lik_name):
    count_fix = 0
    fit_successed,model = fit_GP(count_fix,lik_name)
    
    if fit_successed:
        log_likelihood = model.log_posterior_density().numpy()
        xtest = np.linspace(np.min(X)-.1,np.max(X)+.1,100)[:,None]
        if lik_name == 'Gaussian':
            mean, var = model.predict_y(xtest)
            mean = mean.numpy()
            var = var.numpy()
        else:
            # mean and varicance of samples from posterior predictive distribution 
            mean,var = samples_posterior_predictive_distribution(xtest,model)         
        
        plot(lik_name,model,mean,var)
        ckpt = tf.train.Checkpoint(model= model, step=tf.Variable(1))
        dir_name = 'models/'
        if not os.path.exists(dir_name):
            os.mkdir(dir_name)
        filename = dir_name+genes_name[i]
        ckpt.write(filename)

    else:
        log_likelihood = np.nan  # set to Nan in case of Cholesky decomposition failure or optimization failure 
    return log_likelihood
    
```

```python
genes_log_likelihoods = {}
lik_name = 'Poisson'
column_name = ['log_likelihood']

for i in range(Y.shape[0]): 
    print(i,genes_name[i])
    model_log_likelihood = fit_model(lik_name)
    log_likelihood =  [model_log_likelihood]                           
    genes_log_likelihoods[genes_name[i]] = log_likelihood
  
log_likelihood = pd.DataFrame.from_dict(genes_log_likelihoods, orient='index', columns= column_name)
log_likelihood
```
