This notebook shows how to save GPflow 2 models and load them as check points.

```python
import numpy as np
import random
import tensorflow as tf
import gpflow
import warnings
warnings.filterwarnings('ignore')
%pylab inline 
```

```python
print('tensorflow version:',tf.__version__)
print('GPflow version:',gpflow.__version__)
```

### Sample synthetic data from Gaussian Process

```python
tf.compat.v1.reset_default_graph()
tf.compat.v1.set_random_seed(0)
gpflow.config.set_default_float(np.float64)
np.random.seed(0)
tf.random.set_seed(0)

def sample_GP(ls):
    kern =  gpflow.kernels.RBF( lengthscales = ls,variance = 1.) 
    K = kern(X) + np.eye(N) * 1e-4
    L = np.linalg.cholesky(K)  
    mu = np.ones((N, 1)) 
    f = np.dot(L, np.random.randn(N, 1)) + mu 
    plt.plot(X, f,color= 'green',label = 'latent function f(x)')
    plt.legend()
    plt.show()
    return f

N = 30
X = np.sort(random.randint(0,100,30)).reshape(-1, 1)
lengthscales = [6.,9.]
Y = []
for ls in lengthscales:
    f = sample_GP(ls) 
    Y.append(f)
    plt.plot(X, f,'kx')
    plt.show()
```

### Save GPflow models 

```python
for i in range(len(Y)):

    k = gpflow.kernels.RBF()
    m = gpflow.models.GPR(data=(X.astype(float), Y[i]), kernel=k)
  
    opt = gpflow.optimizers.Scipy()
    opt.minimize(m.training_loss, m.trainable_variables, options=dict(maxiter=100))
    gpflow.utilities.print_summary(m, fmt='notebook')
    ckpt = tf.train.Checkpoint(model=m, step=tf.Variable(1))
    ckpt.write('models/gene-'+str(i+1))
```

### Load GPflow models

```python
k = gpflow.kernels.RBF()
for i in range(2):
    m = gpflow.models.GPR(data=(X.astype(float), Y[i]), kernel=k)
    ckpt = tf.train.Checkpoint(model=m, step=tf.Variable(1))
    ckpt.restore('models/gene-'+str(i+1))
    gpflow.utilities.print_summary(m, fmt='notebook')
```
